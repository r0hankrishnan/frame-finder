```python
from pathlib import Path
from itertools import zip_longest
import pandas as pd
import numpy as np
```

```python
# Define data directory and load data
DATA_DIR = Path().cwd().parent / "data"
cleaned_data = pd.read_csv(DATA_DIR / "interim" / "racquets_cleaned_2026_05_16.csv", index_col=0)
df = cleaned_data.copy()
```

```python
# Display object and numeric columns
obj_cols = df.select_dtypes(include = ["object"]).columns.to_list()
num_cols = df.select_dtypes(include = [np.number]).columns.to_list()

print(
    pd.DataFrame(
        list(zip_longest(obj_cols, num_cols, fillvalue = None)),
        columns = ["Obj Cols", "Num Cols"]
    )
)
```

Pretty much all of the object columns contain useful semantic information. Take `racquet_power_level` for example. This discrete column has the following unique values:

* Low
* Low-Medium
* Medium
* Medium-High
* High

This is a semantically useful feature for a racquet. If a user says "I want a high powered racquet", knowing the value of `racquet_power_level` would come in handy.
