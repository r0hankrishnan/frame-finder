```python
import numpy as np
import pandas as pd
```

```python
# PRICE FILTER
# TYPE = "max" --> Actually doesn't make sense to have a price floor, who is looking for a racquet that's "AT LEAST $200"?
PRICE = 200
```

```python
# Generate 300 random prices 
racquet_prices = np.random.randint(low = 100, high = 400 + 1, size = 300)
```

```python
# Get "deviation" from goal price for each racquet
# We want racquets LOWER than price to have a HIGHER absolute score so
# that their normalized score ADDS to the total relevance score of the racquet
# --> Should do PRICE - racquet_price so that racquets > PRICE are penalized
racquet_price_deviations = PRICE - racquet_prices

racquet_price_deviations
```

```python
# Now need a way to take these deviations and turn them into a normalized "score"

# 1. Min-max normalize deviations so they're between 
norm_price_scores = np.array(list(map(lambda x: (x - racquet_price_deviations.mean()) / (racquet_price_deviations.max() - racquet_price_deviations.min()), racquet_price_deviations)))
```

```python
norm_price_scores.max()
```
